```python
# Group the data by flight duration and calculate the mean CO2 emissions
mean_co2_emissions_by_duration = data.groupby('duration')['co2_emissions'].transform('mean')

# Round the mean CO2 emissions to the nearest whole number
rounded_mean_co2_emissions = mean_co2_emissions_by_duration.round()

# Create a mask to identify null values in co2_emissions
null_mask = data['co2_emissions'].isnull()

# Fill the null values in co2_emissions with the rounded mean CO2 emissions for the corresponding flight duration
data.loc[null_mask, 'co2_emissions'] = rounded_mean_co2_emissions[null_mask]
```

```python
def fill_avg_co2_emissions(data):
    # Create a mask to identify null values in 'avg_co2_emissions_for_this_route'
    null_mask = data['avg_co2_emission_for_this_route'].isnull()

    # Group the data by 'from_country' and 'dest_country' and calculate the mean of 'co2_emissions'
    mean_co2_emissions = data.groupby(['from_country', 'dest_country'])['co2_emissions'].transform('mean')
    
    # Round the calculated mean values
    mean_co2_emissions = mean_co2_emissions.round()

    # Fill null values in 'avg_co2_emissions_for_this_route' with the calculated mean, only for non-null 'co2_emissions'
    data.loc[~data['co2_emissions'].isnull() & null_mask, 'avg_co2_emission_for_this_route'] = mean_co2_emissions

# Usage example:
fill_avg_co2_emissions(data)
```

```python
import numpy as np

def fill_co2_percentage(data):
    # Create a mask to identify null values in 'co2_percentage'
    null_mask = data['co2_percentage'].isnull()

    # Calculate the percentage difference between 'co2_emissions' and 'avg_co2_emission_for_this_route'
    percentage_difference = (data['co2_emissions'] - data['avg_co2_emission_for_this_route']) / data['avg_co2_emission_for_this_route'] * 100

    # Replace null values in 'co2_percentage' with the calculated percentage difference and round to the nearest whole number
    data.loc[null_mask, 'co2_percentage'] = np.round(percentage_difference[null_mask])

# Usage example:
fill_co2_percentage(data)
```
